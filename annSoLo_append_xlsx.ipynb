```python
import numpy as np
import pandas as pd
import glob
import sys
import os
from openpyxl import load_workbook

path = 'C:\\2022-1\\New_Decoy\\BEST\\result\\ANN-SoLo\\NN_result'

def file_input(path):
    i = 0
    
    files = os.listdir(path)
    
    for file in files:
        wb = load_workbook(path + "\\" + file)
        print(file)
        ws = wb.active
        ws.delete_rows(1, 35) # 1번째 row부터 총 35줄 삭제
        wb.save(path + "\\" + file)
        i += 1

file_input(path)
```

```python
# 파일 union
all_data = pd.DataFrame()

for f in glob.glob('C:/2022-1/New_Decoy/BEST/result/ANN-SoLo/NN_result/NN_*.xlsx'):
    df = pd.read_excel(f)
    all_data = all_data.append(df, ignore_index=True)

print(all_data.shape)
```
