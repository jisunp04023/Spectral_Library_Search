```python
f = open("Target_predicted_10.msp")
total = 0
SEQ_T = []

while True:
    line = f.readline()
    
    if not line: break
    
    if line.startswith('Name'):
        total += 1
        if not total % 50000:
            print(total)
        
        seq = line.split(' ')[1].split('/')[0]
        #print(seq)
        SEQ_T.append(seq)
        
f.close()
print(total)

f = open("revDecoy_predicted_10.msp")
total = 0
SEQ_D = []

while True:
    line = f.readline()
    
    if not line: break
    
    if line.startswith('Name'):
        total += 1
        if not total % 50000:
            print(total)
        
        seq = line.split(' ')[1].split('/')[0]
        #print(seq)
        SEQ_D.append(seq)
        
f.close()
print(total)
```

```python
N2_T = []
C2_T = []
C1_T = []

for i in SEQ_T:
    n2 = i[:2]
    c2 = i[-3:-1]
    c1 = i[-2]
    N2_T.append(n2)
    C2_T.append(c2)
    C1_T.append(c1)
    
N2_D = []
C2_D = []
C1_D = []

for i in SEQ_D:
    n2 = i[:2]
    c2 = i[-3:-1]
    c1 = i[-2]
    N2_D.append(n2)
    C2_D.append(c2)
    C1_D.append(c1)
```

```python
from collections import Counter

n2_t = dict(Counter(N2_T))
c2_t = dict(Counter(C2_T))
c1_t = dict(Counter(C1_T))

n2_d = dict(Counter(N2_D))
c2_d = dict(Counter(C2_D))
c1_d = dict(Counter(C1_D))
```
