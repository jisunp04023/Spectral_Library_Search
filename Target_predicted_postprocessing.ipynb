```python
from pprint import pprint

with open("Target_predicted_processed.msp", "w") as f:
    old = open("Target_predicted.msp")
    total = 0

    while True:
        if total == 0:
            line = old.readline()
    
        if not line: break
    
        if line.startswith('Name: '):
            total += 1
            if not total % 10000:
                print(total)
        
            f.write(line)
            line = old.readline()
            f.write(line)
            line = old.readline()
            f.write(line)
            line = old.readline()
            
            line = old.readline()
            
            peaks = []
            
            while True:
                if not line: break
                if line.startswith('Name: '): break
                
                peak = line.split('\t')
                
                if peak[1] != '0.0':
                    peak[0] = float(peak[0])
                    peak[1] = float(peak[1])
                    
                    peaks.append(peak)
                    
                line = old.readline()
            
            peaks.sort()
            
            num = "Num peaks: " + str(len(peaks)) + "\n"
            f.write(num)
            
            for i in peaks:
                p = str(i[0]) + "\t" + str(i[1]) + "\t" + str(i[2])
                f.write(p)
                
old.close()
print(total)
```
