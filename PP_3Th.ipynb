```python
%matplotlib inline
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PPt1_1 = pd.read_excel('./0725/3Th/PP_target1_0.005.xlsx')
PPt1_1.head()
```

```python
PPt1_2 = pd.read_excel('./0725/3Th/PP_target2_0.005.xlsx')
PPt1_2.head()
```

```python
PPd1_1 = pd.read_excel('./0725/3Th/PP_decoy1_0.005.xlsx')
PPd1_1.head()
```

```python
PPd1_2 = pd.read_excel('./0725/3Th/PP_decoy2_0.005.xlsx')
PPd1_2.head()
```

```python
PPt1 = list(PPt1_1['Dot']) + list(PPt1_2['Dot'])
PPt1
```

```python
PPd1 = list(PPd1_1['Dot']) + list(PPd1_2['Dot'])
PPd1
```

```python
ppt1 = plt.hist(PPt1, bins = 50)
ppd1 = plt.hist(PPd1, bins = 50)
```

```python
print(ppt1[1][1] - ppt1[1][0])
print(ppd1[1][1] - ppd1[1][0])

plt.figure(figsize = (10, 7))
plt.xlabel('Dot score')
plt.ylabel('Count')
plt.title('Ptarget + Pdecoy delta > 0.005')
plt.plot(ppt1[1][:-1] + 0.0085, ppt1[0])
plt.plot(ppd1[1][:-1] + 0.0075, ppd1[0])
plt.show()
```

```python
score = 0

for i in range(0, 1000):
    score += 0.001
    print('score: ', score)
    
    nnt = 0
    nnd = 0
    
    for j in PPt1:
        if j >= score:
            nnt += 1
    print('nnt: ', nnt)
    
    for k in PPd1:
        if k >= score:
            nnd += 1
    print('nnd: ', nnd)
    
    print('FDR: ', nnd / nnt)
    print("\n")
    
    if (nnd / nnt) <= 0.01:
        print(nnt)
        print(nnd)
        print(nnd / nnt)
        print(score)
        break
"""
결과
score: 0.6020000000000004

ppt: 160027
ppd: 1575

FDR: 0.009842089147456367
"""
```
