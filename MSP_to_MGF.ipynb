```python
from pprint import pprint

with open("Ntarget.mgf", "w") as f:
    old = open("Ntarget.msp")
    total = 0
    
    while True:
        line = old.readline()
        
        if not line: break
        
        if line.startswith('Name: '):
            total += 1
            if not total % 10000:
                print(total)
            
            t = line.split('/')
            title = t[0] + '/' + t[1][0]
            charge = title[-1] + '+'
            scans = total
            
            #print(title)
            #print(charge)
            #print(scans)
            
            comment = old.readline()
            c = comment.split('Parent=')
            pepmass = c[1].split(' ')[0]
            #print(pepmass)
            
            line = old.readline()
            line = old.readline()
            
            peaks = []
            while True:
                if line.startswith('\n'):
                    break
                if not line: break
                
                peak = line.split('\t')
                peak = peak[:-1]
                peaks.append(peak)
                
                line = old.readline()
            
            #pprint(peaks)
            
            f.write('BEGIN IONS\n')
            f.write('TITLE=' + title + '\n')
            f.write('PEPMASS=' + pepmass + '\n')
            f.write('CHARGE=' + charge + '\n')
            f.write('SCANS=' + str(scans) + '\n')
            
            for i in peaks:
                f.write(i[0] + '\t' + i[1] + '\n')
            
            f.write('END IONS\n\n')
            
    old.close()
print(total)
```
