```python
%matplotlib inline
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
NNt1_1 = pd.read_excel('./0725/3Th/NN_target1_0.005.xlsx')
NNt1_1.head()
```

```python
NNt1_2 = pd.read_excel('./0725/3Th/NN_target2_0.005.xlsx')
NNt1_2.head()
```

```python
NNd1_1 = pd.read_excel('./0725/3Th/NN_decoy1_0.005.xlsx')
NNd1_1.head()
```

```python
NNd1_2 = pd.read_excel('./0725/3Th/NN_decoy2_0.005.xlsx')
NNd1_2.head()
```

```python
NNt1 = list(NNt1_1['Dot']) + list(NNt1_2['Dot'])
NNt1
```

```python
NNd1 = list(NNd1_1['Dot']) + list(NNd1_2['Dot'])
NNd1
```

```python
nnt1 = plt.hist(NNt1, bins = 50)
nnd1 = plt.hist(NNd1, bins = 50)
```

```python
print(nnt1[1][1] - nnt1[1][0])
print(nnd1[1][1] - nnd1[1][0])

plt.figure(figsize = (10, 7))
plt.xlabel('Dot score')
plt.ylabel('Count')
plt.title('Ntarget + Ndecoy delta > 0.005')
plt.plot(nnt1[1][:-1] + 0.0095, nnt1[0])
plt.plot(nnd1[1][:-1] + 0.0085, nnd1[0])
plt.show()
```

```python
score = 0

for i in range(0, 1000):
    score += 0.001
    print('score: ', score)
    
    nnt = 0
    nnd = 0
    
    for j in NNt1:
        if j >= score:
            nnt += 1
    print('nnt: ', nnt)
    
    for k in NNd1:
        if k >= score:
            nnd += 1
    print('nnd: ', nnd)
    
    print('FDR: ', nnd / nnt)
    print("\n")
    
    if (nnd / nnt) <= 0.01:
        print(nnt)
        print(nnd)
        print(nnd / nnt)
        print(score)
        break
"""
결과
score: 0.6670000000000005

nnt: 217994
nnd: 2179

FDR: 0.00999568795471435
"""
```
