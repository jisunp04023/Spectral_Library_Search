```python
%matplotlib inline
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
one = pd.read_excel('three.xlsx')
one.head()
```

```python
Proteins = one['Proteins']
for i in range(1000):
    print(Proteins[i])
```

```python
old = open("Pdecoy_annotated_THIS.msp")
    
idx = 0
total = 0
dic = {}
    
while True:
    oldline = old.readline()
        
    if not oldline: break
        
    if oldline.startswith('Comment: '):
        total += 1
        #if not total % 1000:
            #print(total)
        
        comment = oldline.split('Protein="')
        protein = comment[1].split(' ')[0][:-1]
        #print(protein)
        
        oldline = old.readline()
        numpeak = int(oldline[11:-1])
        #print(numpeak)
        
        dic.update({protein : numpeak})
            
old.close
print(total)
dic
```

```python
numpeaks = []

for i in Proteins:
    if i in dic:
        numpeaks.append(dic[i])
        
numpeaks
```

```python
plt.hist(numpeaks, bins = 35)
plt.title('(1)')
plt.show()
```
