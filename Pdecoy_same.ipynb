```python
from pprint import pprint

with open("Pdecoy_annotated_same_THIS.msp", "w") as f:
    old = open("Pdecoy_annotated_THIS.msp")
    total = 0
    i = -1
    
    while True:
        if total == 0:
            line = old.readline()
        
        if not line: break
            
        if line.startswith('Name: '):
            if not total % 100:
                print(total)
            total += 1
            i += 1
            f.write(line)
            #print(line)
            line = old.readline()
            #f.write(line)
            #print(line)
            line = old.readline()
            f.write(line)
            #print(line)
            line = old.readline()
            
            numpeak_PD = int(line[11:-1])
            
            line = old.readline()
            
            peaks = []
            
            
            while True:
                if not line: break
                
                peak = line.split('\t')
                
                if len(peak) == 1:
                    break
                    
                peak[0] = float(peak[0])
                peak[1] = float(peak[1])
                    
                peaks.append(peak)
                
                line = old.readline()
                
            
            if numpeak_PD > numpeaks_PT[i]:
                peaks.sort(key = lambda x : -x[1])
                peaks = peaks[:numpeaks_PT[i]]
                peaks.sort(key = lambda x : x[0])
            
            f.write("Num peaks: " + str(len(peaks)) + "\n")
            #print("Num peaks: " + str(len(peaks)) + "\n")

            for j in peaks:
                f.write(str(j[0]) + "\t" + str(j[1]) + "\t" + j[2])
                #print(str(j[0]) + "\t" + str(j[1]) + "\t" + j[2])
                
old.close()
print(total)
```
