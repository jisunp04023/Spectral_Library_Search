```python
with open("Ntarget_NEW.msp", "w") as file:
    f = open("Ntarget.msp")
    total = 0

    while True:
        if not total:
            line = f.readline()
        if not line: break
    
        if line.startswith('Name'):
            total += 1
            if not total % 50000:
                print(total)
            
            origin = line
            s = origin.split('/')
            new = s[0] + '/' + s[1][0] + '\n'
            file.write(new)
            
            line = f.readline()
        
            while True:
                if line.startswith('Name'):
                    break
                if not line: break
            
                file.write(line)
                line = f.readline()

    f.close()
print(total)
```
