```python
NNd1 = list(NNd1_1['Dot']) + list(NNd1_2['Dot']) + list(NNd1_3['Dot'])
NNd1
```

```python
nnt1 = plt.hist(NNt1, bins = 50)
nnd1 = plt.hist(NNd1, bins = 50)
```

```python
print(nnt1[1][1] - nnt1[1][0])
print(nnd1[1][1] - nnd1[1][0])

plt.figure(figsize = (10, 7))
plt.xlabel('Dot score')
plt.ylabel('Count')
plt.title('Ntarget + Ndecoy delta > 0.005')
plt.plot(nnt1[1][:-1] + 0.0095, nnt1[0])
plt.plot(nnd1[1][:-1] + 0.009, nnd1[0])
plt.show()
```

```python
score = 0

for i in range(0, 1000):
    score += 0.001
    print('score: ', score)
    
    nnt = 0
    nnd = 0
    
    for j in NNt1:
        if j >= score:
            nnt += 1
    print('nnt: ', nnt)
    
    for k in NNd1:
        if k >= score:
            nnd += 1
    print('nnd: ', nnd)
    
    print('FDR: ', nnd / nnt)
    print("\n")
    
    if (nnd / nnt) <= 0.01:
        print(nnt)
        print(nnd)
        print(nnd / nnt)
        print(score)
        break
"""
결과
score:  0.6090000000000004
nnt:  450468
nnd:  4475
FDR:  0.009934112966958807
"""
```
