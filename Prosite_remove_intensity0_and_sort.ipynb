```python
from pprint import pprint

with open("Target_predicted_expand_10_1f.msp", "w") as f:
    old = open("Target_predicted_expand_10_1.msp")
    total = 0
    
    while True:
        if not total:
            line = old.readline()
        
        if not line: break
        
        if line.startswith('Name: '):
            total += 1
            if not total % 50000:
                print(total)
            
            f.write(line)
            line = old.readline()
            f.write(line)
            line = old.readline()
            f.write(line)
            line = old.readline()
            line = old.readline()
            
            peaks = []
            while True:
                if line.startswith('Name: '):
                    break
                if line.startswith('\n'):
                    break
                if not line: break
                
                peak = line.split('\t')
                peak[0] = float(peak[0])
                peak[1] = float(peak[1])
                peaks.append(peak)
                
                line = old.readline()
            
            #pprint(peaks)
            peaks.sort()
            
            peaks2 = []
            for i in peaks:
                if i[1] != 0.0:
                    peaks2.append(i)
            
            #pprint(peaks2)
            
            f.write('Num peaks: ' + str(len(peaks2)) + '\n')
            
            for i in peaks2:
                f.write(str(i[0]) + '\t' + str(i[1]) + '\t' + i[2])
```
