```python
with open('Ntarget_NEW.msp', 'w') as file:
    f = open('Ntarget.msp')
    total = 0
    
    while True:
        if not total:
            line = f.readline()
        if not line: break
        
        if line.startswith('Name'):
            total += 1
            if not total % 50000:
                print(total)
            
            #print(line[:-1])
            s = line[6:-1].split('/')
            
            seq = s[0]
            
            charge = s[1][0]
            modnum = s[1][2]
            
            name = 'Name: ' + seq + '/' + charge + '\n'
            #print(name)
            file.write(name)
            
            line = f.readline()
            comment = line.split(' ')
            parent = comment[7].split('=')[1]
            collision_energy = comment[9].split('=')[1]
            
            mw = 'MW: ' + parent + '\n'
            #print(mw)
            file.write(mw)
            
            if modnum == '0':
                mods = ' Mods=0'
                modstring = ' ModString=' + seq + '///' + charge
                #print(mods)
                #print(modstring)
                
            
            else:
                modification = []
                
                m1 = s[1][3:]
                m2 = re.split('[(|)]', m1)
                
                for i in m2:
                    tmp = i.split(',')
                    if i != '' and len(tmp) == 3:
                        modification.append(tmp)
                
                #print(modification)
                mods = ' Mods=' + modnum
                for i in modification:
                    mods += '/'
                    mods += i[0]
                    
                    if i[1] == 'M':
                        mods += ',M,Oxidation'
                    
                    elif i[1] == 'C':
                        mods += ',C,Carbamidomethyl'
                        
                #print(mods)
                
                modstring = ' ModString=' + seq + '//'
                cnt = 0
                for i in modification:
                    cnt += 1
                    if i[1] == 'M':
                        modstring =  modstring + 'Oxidation@M' + i[0]
                        
                    elif i[1] == 'C':
                        modstring = modstring + 'Carbamidomethyl@C' + i[0]
                    
                    if cnt != int(modnum):
                        modstring += '; '
                
                #print(modstring)
            
            comment = 'Comment: Parent=' + parent + ' Collision_energy=' + collision_energy + mods + modstring + ' Protein="Target"\n'
            #print(comment)
            file.write(comment)
            
            
            
            line = f.readline()
            
            while True:
                if line.startswith('Name'):
                    break
                if not line: break
                
                #print(line)
                file.write(line)
                line = f.readline()
    
    f.close()
file.close()
print(total)
```
