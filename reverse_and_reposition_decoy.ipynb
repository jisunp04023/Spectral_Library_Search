```python
from pprint import pprint

f = open('D:/Research/msp/Ntarget_filtered_NIST.msp')
total = 0
INTENS = []

while True:
    if not total:
        line = f.readline()
    
    if not line: break
    
    if line.startswith('Name'):
        total += 1
        if not total % 50000:
            print(total)
            
        line = f.readline()
        line = f.readline()
        line = f.readline()
        line = f.readline()
        
        INTEN = []
        while True:
            if not line: break
            if line.startswith('Name'): break
            
            data = line.split('\t')
            inten = data[1]
            annotation = data[2][1:].split('/')[0]
            
            INTEN.append([annotation, inten])
            
            line = f.readline()
            
        #pprint(INTEN)
        INTENS.append(INTEN)
        
        
f.close()
print(total)
```

```python
with open('reverse_and_reposition.msp', 'w') as file:
    f = open('D:/Research/msp/revDecoy_predicted_NIST_mz100_same.msp')
    total = 0

    while True:
        if not total:
            line = f.readline()
    
        if not line: break
    
        if line.startswith('Name'):
            total += 1
            if not total % 50000:
                print(total)
        
            file.write(line)
            line = f.readline()
            file.write(line)
            line = f.readline()
            file.write(line)
            line = f.readline()
            file.write(line)
            line = f.readline()
        
            peak = []
            while True:
                if not line: break
                if line.startswith('Name'): break
            
                data = line.split('\t')
                mz = data[0]
                anno = data[2]
                annotation = data[2][1:].split('/')[0]
            
                inten = 100
                for i in INTENS[total - 1]:
                    if i[0] == annotation:
                        inten = i[1]
                        break
            
                #print([mz, inten, anno])
                peak.append([mz, inten, anno])
            
                line = f.readline()
        
            #pprint(peak)
            for i in peak:
                file.write(i[0] + '\t' + i[1] + '\t' + i[2])
        
    f.close()
file.close()
print(total)
```
