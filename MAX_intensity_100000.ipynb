```python
from pprint import pprint

with open("Target_predicted_10.msp", "w") as f:
    old = open("Target_predicted_Na.msp")
    total = 0
    
    while True:
        if not total:
            line = old.readline()
            
        if not line: break
        
        if line.startswith('Name: '):
            total += 1
            if not total % 50000:
                print(total)
            
            f.write(line)
            line = old.readline()
            f.write(line)
            line = old.readline()
            f.write(line)
            line = old.readline()
            f.write(line)
            line = old.readline()
            
            peaks = []
            while True:
                if line.startswith('Name: '):
                    break
                if line.startswith('\n'):
                    break
                if not line: break
                
                peak = line.split('\t')
                peaks.append(peak)
                
                line = old.readline()
                
            #pprint(peaks)
            max = float(peaks[len(peaks) - 1][0])
            #print(max)
            x = 100000 / max
            #print(x)
            
            for i in peaks:
                i[0] = float(i[0]) * x
                f.write(str(i[0]) + '\t' + i[1] + '\t' + i[2])
                
    old.close()
print(total)
```
