```python
from pprint import pprint

with open("Target_predicted_10.msp", "w") as f:
    old = open("Target_predicted_Na.msp")
    total = 0
    
    while True:
        if not total:
            line = old.readline()
            
        if not line: break
        
        if line.startswith('Name: '):
            total += 1
            if not total % 50000:
                print(total)
            
            f.write(line)
            line = old.readline()
            f.write(line)
            line = old.readline()
            f.write(line)
            line = old.readline()
            f.write(line)
            line = old.readline()
            
            peaks = []
            Max = 0
            while True:
                if line.startswith('Name: '):
                    break
                if line.startswith("\n"):
                    break
                if not line: break
                
                peak = line.split('\t')
                inten = float(peak[1])
                Max = max(Max, inten)
                peaks.append(peak)
                
                line = old.readline()
                
            #pprint(peaks)
            #print(Max)
            x = 100000 / Max
            
            for i in peaks:
                i[1] = float(i[1]) * x
                f.write(i[0] + '\t' + str(i[1]) + '\t' + i[2])
                
    old.close()
print(total)
```
