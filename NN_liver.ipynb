```python
%matplotlib inline
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
NNt1 = pd.read_excel('./delta_cutoff/liver/NN(0.01Th)/NN(0.01Th)_target.xlsx')
NNt1.head()
```

```python
NNd1 = pd.read_excel('./delta_cutoff/liver/NN(0.01Th)/NN(0.01Th)_decoy.xlsx')
NNd1.head()
```

```python
nnt1 = plt.hist(NNt1['Dot'], bins = 50)
nnd1 = plt.hist(NNd1['Dot'], bins = 50)
```

```python
print(nnt1[1][1] - nnt1[1][0])
print(nnd1[1][1] - nnd1[1][0])

plt.figure(figsize = (10, 7))
plt.xlabel('Dot score')
plt.ylabel('Count')
plt.title('Ntarget + Ndecoy delta > 0.005')
plt.plot(nnt1[1][:-1] + 0.0095, nnt1[0])
plt.plot(nnd1[1][:-1] + 0.0085, nnd1[0])
plt.show()
```

```python
score = 0

for i in range(0, 1000):
    score += 0.001
    print('score: ' , score)
    
    nntarget = NNt1['Dot'] >= score
    nnt = NNt1[nntarget]
    
    nndecoy = NNd1['Dot'] >= score
    nnd = NNd1[nndecoy]
    
    print('nnt: ', len(nnt))
    print('nnd: ', len(nnd))
    print('FDR: ', len(nnd) / len(nnt))
    print('\n')
    
    if (len(nnd) / len(nnt)) <= 0.01:
        print(len(nnd))
        print(len(nnt))
        print(len(nnd) / len(nnt))
        print(score)
        break
"""
결과
score:  0.5970000000000004
nnt:  104914
nnd:  1037
FDR:  0.009884286177249938
"""
```
