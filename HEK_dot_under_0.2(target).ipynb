```python
%matplotlib inline
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
hekT = pd.read_excel('./dot_under_0.2/NFRP_HEK_target.xlsx')
hekT.head()
```

```python
hekT1 = hekT.loc[:, ['ID']]
hekT1.head()
```

```python
hekt1 = hekT1.values.tolist()
hekt1
```

```python
hekt1 = np.concatenate(hekt1).tolist()
hekt1
```

```python
seq_len = []
for i in hekt1:
    seq_len.append(len(i) - 2)
seq_len
```

```python
plt.figure(figsize = (10, 7))
plt.xlabel('sequence length')
plt.title('HEK target dot < 0.2: sequence length')
seq_length = plt.hist(seq_len, bins = 20)
```

```python
print(seq_length[1][1] - seq_length[1][0])

plt.figure(figsize = (10, 7))
plt.xlabel('sequence length')
plt.ylabel('Count')
plt.title('HEK target dot < 0.2: sequence length')
plt.plot(seq_length[1][:-1] + 0.6, seq_length[0])
plt.show()
```

```python
ch_state = []
for i in hekt1:
    ch_state.append(int(i[-1]))
ch_state
```

```python
plt.figure(figsize = (10, 7))
plt.title('HEK target dot < 0.2: charge state')
plt.xlabel('charge state')
charge_state = plt.hist(ch_state, bins = 10)
```

```python
c_term = []
for i in hekt1:
    c_term.append(i[-3])
c_term
```

```python
counter1 = Counter(c_term)
counter1
```

```python
plt.figure(figsize = (10, 7))
plt.title('HEK target dot < 0.2: C-term')
plt.bar(counter1.keys(), counter1.values())
plt.show
```

```python
seq = []
for i in hekt1:
    seq.append(i[:-2])
seq
```

```python
amino_acid = []
for i in seq:
    for j in i:
        amino_acid.append(j)
amino_acid
```

```python
counter2 = Counter(amino_acid)
sorted(counter2.items())
```

```python
plt.figure(figsize = (10, 7))
plt.title('HEK target dot < 0.2: amino acid')
plt.bar(counter2.keys(), counter2.values())
plt.show
```
