```python
%matplotlib inline
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
lungT1 = pd.read_excel('./dot_under_0.2/NFRP_lung_target_1.xlsx')
lungT1.head()
```

```python
lungT2 = pd.read_excel('./dot_under_0.2/NFRP_lung_target_2.xlsx')
lungT2.head()
```

```python
lungT3 = pd.read_excel('./dot_under_0.2/NFRP_lung_target_3.xlsx')
lungT3.head()
```

```python
print(len(lungT1))
print(len(lungT2))
print(len(lungT3))
lungT = pd.concat([lungT1, lungT2])
lungT = pd.concat([lungT, lungT3])
print(len(lungT))

lungT = lungT.reset_index(drop=True)
lungT
```

```python
# sequence length
print(len(lungT))

seq1 = [] # 0~10
seq2 = [] # 11~20
seq3 = [] # 21~

for i in range(len(lungT)):
    if not i % 20000:
        print(i)
    ID = lungT.loc[i, 'ID']
    seq = ID[:-2]
    
    if len(seq) >= 0 and len(seq) <=10:
        d = lungT.loc[i, 'Dot']
        seq1.append(d)
    
    elif len(seq) >= 11 and len(seq) <= 20:
        d = lungT.loc[i, 'Dot']
        seq2.append(d)
    
    elif len(seq) >= 21:
        d = lungT.loc[i, 'Dot']
        seq3.append(d)
```

```python
# charge state
print(len(lungT))

cs1 = [] # 2
cs2 = [] # 3
cs3 = [] # 4~

for i in range(len(lungT)):
    if not i % 30000:
        print(i)
    ID = lungT.loc[i, 'ID']
    charge = int(ID[-1])
    
    if charge == 2:
        c = lungT.loc[i, 'Dot']
        cs1.append(c)
    
    elif charge == 3:
        c = lungT.loc[i, 'Dot']
        cs2.append(c)
    
    elif charge >= 4:
        c = lungT.loc[i, 'Dot']
        cs3.append(c)
```

```python
# C-term
print(len(lungT))

ct1 = [] # 2
ct2 = [] # 3
ct3 = [] # 4~

for i in range(len(lungT)):
    if not i % 30000:
        print(i)
    ID = lungT.loc[i, 'ID']
    seq = ID[:-2]
    cterm = seq[-1]
    
    if cterm == 'R':
        ct = lungT.loc[i, 'Dot']
        ct1.append(ct)
    
    elif cterm == 'K':
        ct = lungT.loc[i, 'Dot']
        ct2.append(ct)
    
    else:
        ct = lungT.loc[i, 'Dot']
        ct3.append(ct)
```
