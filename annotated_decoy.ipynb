```python
with open('Pdecoy_annotated.msp', 'w') as f:
    d = open("Pdecoy_use.msp")
    
    while True:
        dline = d.readline()
        
        if not dline: break
        
        if dline.startswith('Comment: '):
            comment = dline.split('Protein="')
            f.write(comment[0] + 'Protein="XXX_' + comment[1])
        else:
            f.write(dline)
```
