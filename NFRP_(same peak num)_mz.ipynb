```python
f = open('NrevDecoy_filtered_NEW.msp')
total = 0
ANNOT = []
NrevDecoy_filtered = []

while True:
    if not total:
        line = f.readline()
    
    if not line: break
    
    if line.startswith('Name'):
        total += 1
        if not total % 50000:
            print(total)
        
        line = f.readline()
        line = f.readline()
        line = f.readline()
        line = f.readline()
        
        annot = []
        peaks = []
        while True:
            if not line: break
            if line.startswith('Name'): break
            
            data = line.split('\t')
            peaks.append(data)
            annotation = data[2][1:]
            i = annotation.split(',')[0]
            
            a = i.split('/')[0]
            if a.startswith('b') or a.startswith('y'):
                if (not '-' in a) and (not '+' in a):
                    annot.append(a)
                    
            line = f.readline()
        
        #print(annot)
        NrevDecoy_filtered.append(peaks)
        ANNOT.append(annot)
            
f.close()
print(total)
```

```python
from pprint import pprint

with open('revDecoy_predicted_NIST_mz100_same.msp', 'w') as file:
    f = open('revDecoy_predicted_NIST.msp')
    total = 0

    while True:
        if not total:
            line = f.readline()
    
        if not line: break
    
        if line.startswith('Name'):
            total += 1
            if not total % 50000:
                print(total)
        
            file.write(line)
        
            line = f.readline()
            file.write(line)
            line = f.readline()
            file.write(line)
            line = f.readline()
            line = f.readline()
        
            peaks = []
            annot = []
            number = []
            NEWpeaks = []
            cnt = 0
            while True:
                if not line: break
                if line.startswith('Name'): break
            
                data = line.split('\t')
                peaks.append(data)
                annotation = data[2][1:]
                i = annotation.split(',')[0]

                a = i.split('/')[0]
                annot.append(a)
                number.append(cnt)
                cnt += 1
            
                line = f.readline()
        
        
            ANNOT_idx = -1
            for j in ANNOT[total - 1]:
                ANNOT_idx += 1
            
                if j in annot: # peak   
                    idx = number[annot.index(j)]
                    NEWpeaks.append(peaks[idx])
                
                else:          # peak   
                    mz = NrevDecoy_filtered[total - 1][ANNOT_idx][0]
                    a = NrevDecoy_filtered[total - 1][ANNOT_idx][2]
                    NEWpeaks.append([mz, 100, a])
        
            file.write('Num peaks: ' + str(len(NEWpeaks)) + '\n')

            for k in NEWpeaks:
                file.write(k[0] + '\t' + str(k[1]) + '\t' + k[2])
        
    f.close()
file.close()
print(total)
```

```python
# intensity 
intensity = []

f = open('revDecoy_predicted_NIST.msp')
total = 0

while True:
    line = f.readline()
    
    if not line: break
    
    if line.startswith('Num'):
        total += 1
        if not total % 50000:
                print(total)
        
        line = f.readline()
        
        while True:
            if not line: break
            if line.startswith('Name'): break
            
            data = line.split('\t')
            inten = float(data[1])
            intensity.append(inten)
            
            line = f.readline()
            
f.close()
print(total)
```
