```python
%matplotlib inline
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from pprint import pprint
```

```python
nistT = pd.read_excel('./target_decoy/HEK(precursor_swap)/peak_match/TPRP_10_012_target.xlsx')
nistT.head()
```

```python
nistD = pd.read_excel('./target_decoy/HEK(precursor_swap)/peak_match/TPRP_10_012_decoy.xlsx')
nistD.head()
```

```python
id_nist_t = nistT['ID'].values.tolist()
id_nist_t
```

```python
id_nist_d = nistD['ID'].values.tolist()
id_nist_d
```

```python
SEQ_T = []
SEQ_D = []

for i in id_nist_t:
    seq = ''
    for j in i:
        if j == '[' or j == ']' or j == '1' or j == '6' or j == '0' or j == '4' or j == '7':
            continue
        else:
            seq += j
    seq = seq.split('/')[0]
    SEQ_T.append(seq)
print(len(SEQ_T))
    
for i in id_nist_d:
    seq = ''
    for j in i:
        if j == '[' or j == ']' or j == '1' or j == '6' or j == '0' or j == '4' or j == '7':
            continue
        else:
            seq += j
    seq = seq.split('/')[0]
    SEQ_D.append(seq)
print(len(SEQ_D))
```

```python
N2_T = []
C2_T = []
C1_T = []

for i in SEQ_T:
    n2 = i[:2]
    c2 = i[-3:-1]
    c1 = i[-2]
    N2_T.append(n2)
    C2_T.append(c2)
    C1_T.append(c1)
    
N2_D = []
C2_D = []
C1_D = []

for i in SEQ_D:
    n2 = i[:2]
    c2 = i[-3:-1]
    c1 = i[-2]
    N2_D.append(n2)
    C2_D.append(c2)
    C1_D.append(c1)
```

```python
from collections import Counter

n2_t = dict(Counter(N2_T))
c2_t = dict(Counter(C2_T))
c1_t = dict(Counter(C1_T))

n2_d = dict(Counter(N2_D))
c2_d = dict(Counter(C2_D))
c1_d = dict(Counter(C1_D))
```
