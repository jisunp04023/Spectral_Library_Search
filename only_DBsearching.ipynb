```python
%matplotlib inline
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
t = pd.read_excel('./target_decoy/HEK/TPRP_10_target.xlsx')
t.head() # spectraST TARGET HIT
```

```python
fdrT = t[t['Dot'] > 0.505]
print(len(fdrT))
fdrT.head() # spectraST FDR 1%
```

```python
DB = pd.read_table('./chick_msgf_10ppm_fdr1.tsv')
print(len(DB))
DB.head() # DB Search FDR 1%
```

```python
spectraST = fdrT['### Query'].to_list()
spectraST # spectraST FDR 1% spectrum name
```

```python
ST = []
for i in spectraST:
    s = i.split('.')
    s2 = s[0] + '.' + s[1]
    ST.append(s2)
ST # spectraST FDR 1% spectrum name
```

```python
onlyDB = DB[~DB['Title'].isin(ST)]
print(len(onlyDB))
onlyDB.head() # (DB Search FDR 1%) - (spectraST fdr 1%) = only DB
```

```python
onlyDB_spectrum = onlyDB['Title'].to_list()
onlyDB_charge = onlyDB['Charge'].to_list()
onlyDB_spectrum # spectraST와 형태 맞추기
```

```python
only_DB = []
for i in range(len(onlyDB_charge)):
    s = onlyDB_spectrum[i].split('.')
    s2 = s[0] + '.' + s[1] + '.' + s[1] + '.' + str(onlyDB_charge[i])
    only_DB.append(s2)
only_DB # (DB Search FDR 1%) - (spectraST fdr 1%) = only DB

# only DB spectrum name
```

```python
len(only_DB)
```

```python
spectraST_all = pd.read_excel('./target_decoy/HEK/TPRP_10.xlsx')
spectraST_all.head() # spectraST TARGET + DECOY HIT
```

```python
final = spectraST_all[spectraST_all['### Query'].isin(only_DB)]
print(len(final))
final.head() # only DB로 서치된 것 중 "spectrST로 id된 전체 결과"에 존재하는 것
```

```python
final.to_excel('C:/2022-2/New_Decoy/BEST/result/HEK/onlyDB(TPRP_10).xlsx', header=True, index=False)
```

```python
score = 0

for i in range(0, 1000):
    score += 0.001
    print('score: ' , score)
    
    nntarget = NNt1['Dot'] >= score
    nnt = NNt1[nntarget]
    
    nndecoy = NNd1['Dot'] >= score
    nnd = NNd1[nndecoy]
    
    print('nnt: ', len(nnt))
    print('nnd: ', len(nnd))
    print('FDR: ', len(nnd) / len(nnt))
    print('\n')
    
    if (len(nnd) / len(nnt)) <= 0.01:
        print(len(nnd))
        print(len(nnt))
        print(len(nnd) / len(nnt))
        print(score)
        break
"""
결과
안나옴
"""
```
