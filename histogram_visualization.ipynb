```python
same_target = pd.read_excel('PP_same_target.xlsx')
same_target.head()
```

```python
same_decoy = pd.read_excel('PP_same_decoy.xlsx')
same_decoy.head()
```

```python
plt.hist(same_target['Dot'], bins = 50, alpha = 0.8)
plt.hist(same_decoy['Dot'], bins = 50, alpha = 0.8)
plt.xlabel('Dot score')
plt.ylabel('Count')
plt.title('Ptarget + Pdecoy(same) result')
plt.show()
```

```python
# Ptarget + Pdecoy(same) FDR 1% 구하기

score2 = 0

for i in range(0, 1000):
    score2 += 0.001
    print('score: ' , score2)
    
    pptarget = same_target['Dot'] >= score2
    ppt = same_target[pptarget]
    
    ppdecoy = same_decoy['Dot'] >= score2
    ppd = same_decoy[ppdecoy]
    
    print('nnt: ', len(ppt))
    print('nnd: ', len(ppd))
    print('FDR: ', len(ppd) / len(ppt))
    print('\n')
    
    if (len(ppd) / len(ppt)) <= 0.01:
        print(len(ppd))
        print(len(ppt))
        print(len(ppd) / len(ppt))
        print(score2)
        break
        
"""
결과
score:  0.5140000000000003

target:  188934
decoy:  1860

FDR:  0.009844707675696274
"""
```

```python
plt.hist(NN_target, bins = 50, alpha = 0.8)
plt.hist(NN_decoy, bins = 50, alpha = 0.8)
plt.xlabel('Dot score')
plt.ylabel('Count')
plt.title('SpectraST(3TH): Ntarget + Ndecoy result')
plt.show()
```

```python
score = 0

for i in range(0, 1000):
    score += 0.001
    print('score: ', score)
    
    nnt = 0
    nnd = 0
    
    for j in NN_target:
        if j >= score:
            nnt += 1
    print('nnt: ', nnt)
    
    for k in NN_decoy:
        if k >= score:
            nnd += 1
    print('nnd: ', nnd)
    
    print('FDR: ', nnd / nnt)
    print('\n')
    
    if (nnd / nnt) <= 0.01:
        print(score)
        break
        
"""
결과
score:  0.670
nnt:  216158
nnd:  2106
FDR:  0.009742873268627578
"""
```
