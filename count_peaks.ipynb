```python
%matplotlib inline
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

o = open("Ntarget_use_THIS.msp")

for i in range(1000):
    c = 0
    line = o.readline()
    print(line)

o.close()
```

```python
file = open("Ntarget_use_THIS.msp")

numpeaks_NT = []

while True:
    line = file.readline()
    
    if not line: break
    
    if line.startswith('Num peaks: '):
        numpeaks_NT.append(int(line[11:-1]))

file.close()
numpeaks_NT
```

```python
plt.hist(numpeaks_NT, bins = 50)
plt.title('Ntarget: number of peaks')
plt.show()
```

```python
plt.hist(numpeaks_NT, bins = 200)
plt.xlim([0, 500])
plt.title('Ntarget: number of peaks')
plt.show()
```
