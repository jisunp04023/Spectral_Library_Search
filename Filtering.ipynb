```python
from pprint import pprint

with open("Ptarget_use_THIS.msp", "w") as f:
    old = open("Ntarget_use_THIS.msp")
    total = 0
    
    while True:
        line = old.readline()
    
        if not line: break
        
        if line.startswith('Name: '):
            total += 1
            name = line
            comment = old.readline()
            
            print(name)
            f.write(name)
            print(comment)
            f.write(comment)
            
            line = old.readline()
            line = old.readline()
            
            peaks = []
            while True:
                if line.startswith("\n"):
                    break
                if not line: break
                    
                peak = line.split('\t')
                
                if peak[2][1] == 'y' or peak[2][1] == 'b':
                    if peak[2][2:].split('/')[0].isdigit():
                        print(peak)
                        peaks.append(peak)
                
                line = old.readline()
            print(peaks)
            
            f.write("Num peaks: " + str(len(peaks)) + "\n")
            
            for i in peaks:
                f.write(str(i[0]) + "\t" + str(i[1]) + "\t" + i[2])
                
            
    old.close()
```
