```python
from pprint import pprint
import statistics

f = open("Target_predicted_10.msp")
total = 0

SUM = [] # 합
MED = [] # 중간값
MEAN = [] # 평균값
NUM = [] # 개수

while True:
    line = f.readline()
    
    if not line: break
    
    if line.startswith('MW'):
        total += 1
        if not total % 50000:
            print(total)
        
        line = f.readline()
        line = f.readline()
        line = f.readline()
        
        peaks = []
        while True:
            if line.startswith('Name: '):
                break
            if line.startswith("\n"):
                break
            if not line: break
            
            peak = line.split('\t')
            peaks.append(float(peak[0]))
            
            line = f.readline()
            
        #pprint(peaks)
        
        Sum = 0
        for i in peaks:
            Sum += i
        
        Med = statistics.median(peaks)
        
        Num = len(peaks)
        
        Mean = Sum / Num
        
        #print(Sum)
        #print(Med)
        #print(Mean)
        #print(Num)
        #print("\n")
        
        SUM.append(Sum)
        MED.append(Med)
        MEAN.append(Mean)
        NUM.append(Num)
        
f.close()
print(total)
```

```python
from pprint import pprint
import statistics

f = open("Ntarget_filtered_10.msp")
total = 0

SUM3 = [] # 합
MED3 = [] # 중간값
MEAN3 = [] # 평균값
NUM3 = [] # 개수

while True:
    line = f.readline()
    
    if not line: break
    
    if line.startswith('Comment'):
        total += 1
        if not total % 50000:
            print(total)
        
        line = f.readline()
        line = f.readline()
        
        peaks = []
        while True:
            if line.startswith('Name: '):
                break
            if line.startswith("\n"):
                break
            if not line: break
            
            peak = line.split('\t')
            peaks.append(float(peak[0]))
            
            line = f.readline()
            
        #pprint(peaks)
        
        if not len(peaks):
             continue
                
        Sum = 0
        for i in peaks:
            Sum += i
        
        Med = statistics.median(peaks)
        
        Num = len(peaks)
        
        Mean = Sum / Num
        
        #print(Sum)
        #print(Med)
        #print(Mean)
        #print(Num)
        #print("\n")
        
        SUM3.append(Sum)
        MED3.append(Med)
        MEAN3.append(Mean)
        NUM3.append(Num)
        
f.close()
print(total)
```
