```python
score = 0

for i in range(0, 1000):
    score += 0.001
    print('score: ' , score)
    
    nntarget = t1['Dot'] >= score
    nnt = t1[nntarget]
    
    nndecoy = d1['Dot'] >= score
    nnd = d1[nndecoy]
    
    print('nnt: ', len(nnt))
    print('nnd: ', len(nnd))
    print('FDR: ', len(nnd) / len(nnt))
    print('\n')
    
    if (len(nnd) / len(nnt)) <= 0.01:
        print(len(nnd))
        print(len(nnt))
        print(len(nnd) / len(nnt))
        print(score)
        break
"""
결과
score:  0.489

nnt:  197943
nnd:  1961

FDR:  0.009906892388212768
"""
```
