```python
%matplotlib inline
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pptarget = pd.read_excel('PP_ANN_1_target.xlsx')
pptarget.head()
```

```python
ppdecoy = pd.read_excel('PP_ANN_1_decoy.xlsx')
ppdecoy.head()
```

```python
pptarget2 = pd.read_excel('PP_ANN_2_target.xlsx')
pptarget2.head()
```

```python
ppdecoy2 = pd.read_excel('PP_ANN_2_decoy.xlsx')
ppdecoy2.head()
```

```python
PP_target = list(pptarget['search_engine_score[1]']) + list(pptarget2['search_engine_score[1]'])
PP_target
```

```python
PP_decoy = list(ppdecoy['search_engine_score[1]']) + list(ppdecoy2['search_engine_score[1]'])
PP_decoy
```

```python
score = 0

for i in range(0, 1000):
    score += 0.001
    print('score: ', score)
    
    ppt = 0
    ppd = 0
    
    for j in PP_target:
        if j >= score:
            ppt += 1
    print('ppt: ', ppt)
    
    for k in PP_decoy:
        if k >= score:
            ppd += 1
    print('ppd: ', ppd)
    
    print('FDR: ', ppd / ppt)
    print("\n")
    
    if (ppd / ppt) <= 0.01:
        print(ppt)
        print(ppd)
        print(ppd / ppt)
        print(score)
        break
"""
결과
score: 0.378

ppt: 168336
ppd: 1658

FDR: 0.009849348921205208
"""
```
