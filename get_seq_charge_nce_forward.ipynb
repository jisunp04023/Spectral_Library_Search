```python
best = open("human_hcd_tryp_best.msp")
c = 0

while True:
    bestline = best.readline()
    
    if not bestline: break
    
    if bestline.startswith('Name: '):
        if not (c % 10): print("row ", c, " done")
        c += 1
        
        name = bestline[6:-1] # AAAAAAAAEQQSSNGPVK/2_1(0,A,Acetyl)
        
        s = name.split('/')
        seq = s[0]            # AAAAAAAAEQQSSNGPVK
        charge = s[1][0]      # 2
        
        # modification 있다면?
        if name[-1] != '0':
            
            # ALSVQASLMTPVQPMLAEACK/2_3(8,M,Oxidation)(14,M,Oxidation)(19,C,CAM)
            idx = []  # [8, 14]
            for i in range(0, len(name)):
                if name[i] == '(':
                    mod = ''       # 8,M,Oxidation
                    
                    for j in range(i+1, len(name)):
                        if name[j] == ')': break
                        
                        mod += name[j]
                    
                    m = mod.split(',') # ['8', 'M', 'Oxidation']
                    
                    if m[2] == 'Oxidation':
                        idx += [int(m[0])]            
            
            # idx = [8, 14]
            for i in range(0, len(idx)):
                # 다음 인덱스들에 +4
                for j in range(i+1, len(idx)):
                    idx[j] += 4
                
                # 지금 인덱스 자리에 '(ox)' 추가
                idx[i] += 1
                seq = seq[:idx[i]] + '(ox)' + seq[idx[i]:]
            
            #print('modified: ' + seq)
            
        print(seq + '\t' + charge)
    
    
    # collision energy 찾기
    if bestline.startswith('Comment: '):
        pos_nce = bestline.find('NCE=')
        nce = float(bestline[pos_nce+4 : pos_nce+9])
        print(nce)
        
        

best.close()
```
