```python
f = open("Ntarget_filtered.msp")

total = 0

NP = []

while True:
    line = f.readline()
    
    if not line: break
    
    if line.startswith('Num peaks: '):
        total += 1
        numpeaks = line.split(': ')[1][:-1]
        NP.append(numpeaks)

f.close()

print(total)
NP
```

```python
from pprint import pprint
import random

with open("revDecoy_predicted_same_numpeaks(random).msp", "w") as f:
    old = open("revDecoy_predicted22.msp")
    total = 0
    
    while True:
        if not total:
            line = old.readline()
        
        if not line: break
        
        if line.startswith('Name: '):
            if not total % 10000:
                print(total)
            total += 1
            
            f.write(line)
            line = old.readline()
            f.write(line)
            line = old.readline()
            f.write(line)
            
            line = old.readline()
            DP = int(line.split(': ')[1][:-1])
            if DP > int(NP[total - 1]):
                f.write('Num peaks: ' + NP[total - 1] + '\n')
            else:
                f.write(line)
                
            line = old.readline()
            
            peaks = []
            while True:
                if line.startswith('Name: '):
                    break
                if line.startswith('\n'):
                    break
                if not line: break
                
                peak = line.split('\t')
                peak[0] = float(peak[0])
                peaks.append(peak)
                
                line = old.readline()
                
            #pprint(peaks)
            
            if DP > int(NP[total - 1]):
                num = int(NP[total - 1])
                random.shuffle(peaks)
                peaks = peaks[:num]
                #pprint(peaks)
                peaks.sort()
                #pprint(peaks)
            
            for i in peaks:
                f.write(str(i[0]) + '\t' + i[1] + '\t' + i[2])
                
    old.close()
print(total)
```
