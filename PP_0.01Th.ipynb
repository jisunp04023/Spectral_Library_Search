```python
%matplotlib inline
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PPt1 = pd.read_excel('./0725/0.01Th/PP_target_0.005.xlsx')
PPt1.head()
```

```python
PPd1 = pd.read_excel('./0725/0.01Th/PP_decoy_0.005.xlsx')
PPd1.head()
```

```python
ppt1 = plt.hist(PPt1['Dot'], bins = 50)
ppd1 = plt.hist(PPd1['Dot'], bins = 50)
```

```python
print(ppt1[1][1] - ppt1[1][0])
print(ppd1[1][1] - ppd1[1][0])

plt.figure(figsize = (10, 7))
plt.xlabel('Dot score')
plt.ylabel('Count')
plt.title('Ptarget + Pdecoy delta > 0.005')
plt.plot(ppt1[1][:-1] + 0.0085, ppt1[0])
plt.plot(ppd1[1][:-1] + 0.0075, ppd1[0])
plt.show()
```

```python
score = 0

for i in range(0, 1000):
    score += 0.001
    print('score: ' , score)
    
    nntarget = PPt1['Dot'] >= score
    nnt = PPt1[nntarget]
    
    nndecoy = PPd1['Dot'] >= score
    nnd = PPd1[nndecoy]
    
    print('nnt: ', len(nnt))
    print('nnd: ', len(nnd))
    print('FDR: ', len(nnd) / len(nnt))
    print('\n')
    
    if (len(nnd) / len(nnt)) <= 0.01:
        print(len(nnd))
        print(len(nnt))
        print(len(nnd) / len(nnt))
        print(score)
        break
"""
결과
score:  0.5260000000000004

ppt:  184731
ppd:  1818

FDR:  0.009864298077601315
"""
```
