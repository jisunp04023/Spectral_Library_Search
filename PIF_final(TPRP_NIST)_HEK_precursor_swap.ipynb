```python
MAP = []

for i in range(0, 24):
    file = 'C:/2022-2/New_Decoy/BEST/PEAK/HEK_data/precursor_swap/' + file_name[i] + '.mgf'
    print(file)
    
    f = open(file)
    name = file.split('/')[7][:-4]
    total = 0

    while True:
        line = f.readline()
    
        if not line: break
    
        if line.startswith('BEGIN'):
            total += 1
            if not total % 10000:
                print(total)
        
            line = f.readline()
            scan = int(line.split('=')[1][:-1])
            #print(scan)
        
            line = f.readline()
            line = f.readline()
            line = f.readline()
            line = f.readline()
            cs = int(line.split('=')[1][:-2])
            #print(cs)
        
            data = [name, scan, cs, str(total).zfill(5)]
            MAP.append(data)
        
    f.close()
    print(total)
```
