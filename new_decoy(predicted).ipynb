```python
from pprint import pprint

with open("new_Pdecoy.msp", "w") as f: # Pdecoy 이상한 부분 바꾼! (바로 사용할 수 있는 형태)
    old = open("Pdecoy.msp") # Prosit에서 나온 Pdecoy
    origin = open("human_hcd_tryp_best_decoy.msp")
    
    total = 0

    while True:
        if total == 0:
            oldline = old.readline()
        
        if not oldline: break
        
        if oldline.startswith('Name: '):
            name = oldline[:-1]
            print("oldline----> " + name)
            total += 1
            print(total)
            
            originline = origin.readline()
            
            while True:
                if originline.startswith(name):
                    print("newline----------> " + originline[:-1])
                    f.write(originline)
                    originline = origin.readline()
                    f.write(originline)
                    originline = origin.readline()
                    f.write(originline)
                    
                    oldline = old.readline()
                    oldline = old.readline()
                    oldline = old.readline()
                    oldline = old.readline()
                    
                    print('just checking: ' + oldline)
                    
                    peaks = []
                    while True:
                        if oldline.startswith("Name: "):
                            break
                        if not oldline: break
                        
                        peak = oldline.split('\t') # ['147.11280822753906', '0.07357145845890045', '"y1/0.0ppm"\n']
                        
                        if peak[1] != '0.0':
                            peak[0] = float(peak[0])
                            peak[1] = float(peak[1]) * 100000
                            peaks.append(peak)
                        
                        oldline = old.readline()
                    
                    peaks.sort(key=lambda x:x[0])
                    
                    f.write("Num peaks: " + str(len(peaks)) + "\n")
                    
                    for i in peaks:
                        f.write(str(i[0]) + "\t" + str(i[1]) + "\t" + i[2])
                    
                    break
                    
                else:
                    originline = origin.readline()

print(total)
```
