```python
numpeak = []

f = open('Ntarget_filtered_NIST.msp')
total = 0

while True:
    line = f.readline()
    
    if not line: break
    
    if line.startswith('Num peak'):
        total += 1
        
        numpeak.append(int(line[11:-1]))

f.close()
print(total)
```

```python
# target peak    

from pprint import pprint
import random

with open('revDecoy_predicted_NIST_random_same.msp', 'w') as file:
    f = open('revDecoy_predicted_NIST.msp')
    total = 0

    while True:
        if not total:
            line = f.readline()
    
        if not line: break
    
        if line.startswith('Name'):
            total += 1
            if not total % 50000:
                print(total)
        
            file.write(line)
            line = f.readline()
            file.write(line)
            line = f.readline()
            file.write(line)
            line = f.readline()
            file.write('Num peaks: ' + str(numpeak[total - 1]) + '\n')
            line = f.readline()
        
            peaks = []
        
            while True:
                if not line: break
                if line.startswith('Name'): break
            
                data = line.split('\t')
                data[0] = float(data[0])
                peaks.append(data)
            
                line = f.readline()
            
            #pprint(peaks)
            random.shuffle(peaks)
        
            newpeaks = peaks[:numpeak[total - 1]]

            newpeaks.sort(key = lambda x : x[0])

        
            for i in newpeaks:
                file.write(str(i[0]) + '\t' + i[1] + '\t' + i[2])

    f.close()
file.close()
print(total)
```

```python
# target peak   intensity   

from pprint import pprint
import random

with open('revDecoy_predicted_NIST_upper_same.msp', 'w') as file:
    f = open('revDecoy_predicted_NIST.msp')
    total = 0

    while True:
        if not total:
            line = f.readline()
    
        if not line: break
    
        if line.startswith('Name'):
            total += 1
            if not total % 50000:
                print(total)
        
            file.write(line)
            line = f.readline()
            file.write(line)
            line = f.readline()
            file.write(line)
            line = f.readline()
            file.write('Num peaks: ' + str(numpeak[total - 1]) + '\n')
            line = f.readline()
        
            peaks = []
        
            while True:
                if not line: break
                if line.startswith('Name'): break
            
                data = line.split('\t')
                data[0] = float(data[0])
                data[1] = float(data[1])
                peaks.append(data)
            
                line = f.readline()
            
            #pprint(peaks)
            peaks.sort(key = lambda x : -x[1])
        
            newpeaks = peaks[:numpeak[total - 1]]

            newpeaks.sort(key = lambda x : x[0])

        
            for i in newpeaks:
                file.write(str(i[0]) + '\t' + str(i[1]) + '\t' + i[2])

    f.close()
file.close()
print(total)
```
