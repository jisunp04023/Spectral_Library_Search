```python
NNt3 = NNt1['search_engine_score[1]']
NNd3 = NNd1['search_engine_score[1]']

score = 0

for i in range(0, 1000):
    score += 0.001
    print('score: ', score)
    
    nnt = 0
    nnd = 0
    
    for j in NNt3:
        if j >= score:
            nnt += 1
    print('nnt: ', nnt)
    
    for k in NNd3:
        if k >= score:
            nnd += 1
    print('nnd: ', nnd)
    
    print('FDR: ', nnd / nnt)
    print("\n")
    
    if (nnd / nnt) <= 0.01:
        print(nnt)
        print(nnd)
        print(nnd / nnt)
        print(score)
        break
"""
ê²°ê³¼
score:  0.2880000000000002
nnt:  282784
nnd:  2779
FDR:  0.009827288672626456
"""
```
