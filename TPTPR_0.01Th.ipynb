```python
score = 0

for i in range(0, 1000):
    score += 0.001
    print('score: ', score)
    
    nnt = 0
    nnd = 0
    
    for j in NNt1:
        if j >= score:
            nnt += 1
    print('nnt: ', nnt)
    
    for k in NNd1:
        if k >= score:
            nnd += 1
    print('nnd: ', nnd)
    
    print('FDR: ', nnd / nnt)
    print("\n")
    
    if (nnd / nnt) <= 0.01:
        print(nnt)
        print(nnd)
        print(nnd / nnt)
        print(score)
        break
"""
ê²°ê³¼
score:  0.4910000000000004
nnt:  265090
nnd:  2633
FDR:  0.009932475762948433
"""
```
