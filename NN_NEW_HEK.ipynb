```python
NNt1_2 = pd.read_excel('./target_decoy/HEK/NN_NEW_2_target.xlsx')
NNt1_2.head()
```

```python
NNt3 = list(NNt1_1['Dot']) + list(NNt1_2['Dot'])
NNt3
```

```python
NNd3 = list(NNd1_1['Dot']) + list(NNd1_2['Dot'])
NNd3
```

```python
nnt3 = plt.hist(NNt3, bins = 50)
nnd3 = plt.hist(NNd3, bins = 50)
```

```python
print(nnt3[1][1] - nnt3[1][0])
print(nnd3[1][1] - nnd3[1][0])

plt.figure(figsize = (10, 7))
plt.xlabel('Dot score')
plt.ylabel('Count')
plt.title('NN_NEW: HEK')
plt.plot(nnt3[1][:-1] + 0.0095, nnt3[0])
plt.plot(nnd3[1][:-1] + 0.0085, nnd3[0])
plt.show()
```

```python
print(len(NNt3))
print(len(NNd3))
```

```python
score = 0

for i in range(0, 1000):
    score += 0.001
    print('score: ', score)
    
    nnt = 0
    nnd = 0
    
    for j in NNt3:
        if j >= score:
            nnt += 1
    print('nnt: ', nnt)
    
    for k in NNd3:
        if k >= score:
            nnd += 1
    print('nnd: ', nnd)
    
    print('FDR: ', nnd / nnt)
    print("\n")
    
    if (nnd / nnt) <= 0.01:
        print(nnt)
        print(nnd)
        print(nnd / nnt)
        print(score)
        break
"""
ê²°ê³¼
score:  0.6010000000000004
nnt:  309935
nnd:  3074
FDR:  0.009918208656653814
"""
```
